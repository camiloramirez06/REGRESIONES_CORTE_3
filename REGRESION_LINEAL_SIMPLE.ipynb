```python
# prompt: haz un analisis de regresion lineal simple que la variable X sea "Empleados" y la variable Y sea "Tasa_Crecimiento" y determine en R cuadrado. sin entrenamiento ni validacion

import pandas as pd
import statsmodels.api as sm

# Load the Excel file into a pandas DataFrame.
df = pd.read_excel("/content/drive/MyDrive/CORTE 2/datos_industrias REGRESION.xlsx")

# Define the independent variable (X) and the dependent variable (Y).
X = df['Empleados']
Y = df['Tasa_Crecimiento']

# Add a constant to the independent variable (required for statsmodels).
X = sm.add_constant(X)

# Create a linear regression model.
model = sm.OLS(Y, X).fit()

# Print the model summary, which includes R-squared.
print(model.summary())
```

```python
# prompt: necesito un analisis de regresion lineal simple que la variable X sea Salario y la variable Y sea Tasa_Crecimiento

# Define the independent variable (X) and the dependent variable (Y).
X = df['Salario']
Y = df['Tasa_Crecimiento']

# Add a constant to the independent variable (required for statsmodels).
X = sm.add_constant(X)

# Create a linear regression model.
model = sm.OLS(Y, X).fit()

# Print the model summary, which includes R-squared.
print(model.summary())
```

```python
# prompt: dame su respectivo grafico

import matplotlib.pyplot as plt
import seaborn as sns

# Define the independent variable (X) and the dependent variable (Y).
X_empleados = df['Empleados']
X_salario = df['Salario']
Y = df['Tasa_Crecimiento']

# Create scatter plots
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
sns.scatterplot(x=X_empleados, y=Y)
plt.title('Tasa de Crecimiento vs Empleados')
plt.xlabel('Empleados')
plt.ylabel('Tasa de Crecimiento')

plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
sns.scatterplot(x=X_salario, y=Y)
plt.title('Tasa de Crecimiento vs Salario')
plt.xlabel('Salario')
plt.ylabel('Tasa de Crecimiento')

plt.tight_layout()
plt.show()
```
